```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array(list(range(0, 100))).reshape(-1,1)
Y = np.asarray(X%2==0, dtype='uint8')
plt.plot(X, Y)
```

```python
X=[]
n = len(bin(100)[2:])
for i in range(100):
    b = bin(i)[2:].zfill(n)
    l = [int(x) for x in b]
    X.append(l)

X = np.array(X)
```

```python
X
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron(random_state=42)
p.fit(X, Y.ravel())
```

```python
print(p.score(X, Y))
```
